```python
#draw seperately
plt.boxplot(df['petal_width'])
plt.boxplot(df['petal_length'])
plt.boxplot(df['sepal_length'])
plt.boxplot(df['petal_width'])
```

```python
fig,axes=plt.subplots(2,2)
sns.boxplot(data= df,x='petal_width',ax=axes[0,0])
sns.boxplot(data= df,x='petal_length',ax=axes[1,0])
sns.boxplot(data= df,x='sepal_width',ax=axes[0,1])
sns.boxplot(data= df,x='sepal_length',ax=axes[1,1])
```
