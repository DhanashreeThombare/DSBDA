```python
#deal with skewness
df['age']=df['age'].apply(lambda x:np.log(x) if x>0 else 0)
df['age']
```

```python
q1=df['age'].quantile(0.25)
q3=df['age'].quantile(0.75)

IQR=q3-q1
high= q3+1.5*IQR
low=q1-1.5*IQR

#remove skew
df['age']=df['age'].clip(lower=low,upper=high)
```

```python
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
IQR = q3 - q1
high = q3 + 1.5 * IQR
low = q1 - 1.5 * IQR

# Clip the data to remove outliers
df['age'] = df['age'].clip(lower=low, upper=high)

print("Clipped 'age' column:")
print(df['age'].skew())

```
